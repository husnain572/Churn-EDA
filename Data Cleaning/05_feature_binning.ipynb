# Feature Binning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['CustomerId','Surname','RowNumber'],axis=1,inplace=True)
```

```python
df.head()
```

```python
print(df.Age.min())
print(df.Age.max())
```

```python
labels=['0-20','21-40','41-60','Above 61']
bins=[0,20,40,60,100]
df['Age_bins']=pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
df.head()
```

```python
print(df[['Age',"Age_bins"]])
```

```python
df.Age_bins.value_counts()
```

```python
#making the bar chart on the data
plt.bar(labels,df.Age_bins.value_counts())

plt.title("Age Count")
plt.xlabel("Age Bins")
plt.ylabel("Age Counts")
plt.show()
```
